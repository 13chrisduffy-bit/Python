
# Introduction to Regular Expressions (Regex) in Python

## Step 1 — What is Regex?

A Regular Expression (Regex) is a pattern used to search, match, or extract text.

Think of it as:
- “Find all phone numbers”
- “Check if an email is valid”
- “Extract all dates”
- “Find every word starting with A”

In Python, Regex is handled using the built-in `re` library.


## Step 2 — Import the Regex Library

```python
import re
```

## Step 3 — Find a Word

```python
import re

text = "Python is awesome"

result = re.search("Python", text)

print(result)
```

## Step 4 — Check if Something Was Found

```python
import re

text = "Python is awesome"

result = re.search("Java", text)

print(result)

if result:
    print("Match found")
else:
    print("No match")
```

## Step 5 — Match Any Digit

```python
import re

text = "My number is 42"

result = re.search(r"\d", text)

print(result)
```

## Step 6 — Find All Digits

```python
import re

text = "My numbers are 42 and 99"

result = re.findall(r"\d", text)

print(result)
```

## Step 7 — Match Multiple Digits

```python
import re

text = "My numbers are 42 and 99"

result = re.findall(r"\d+", text)

print(result)
```

## Step 8 — Learn the Most Common Patterns

```python
patterns = {
    r"\d": "Digit",
    r"\D": "Not a digit",
    r"\w": "Letter/number/underscore",
    r"\W": "Not a word character",
    r"\s": "Whitespace",
    r"\S": "Not whitespace",
    ".": "Any character",
    "+": "One or more",
    "*": "Zero or more",
    "?": "Optional",
    "^": "Start of string",
    "$": "End of string"
}

for k, v in patterns.items():
    print(f"{k} -> {v}")
```

## Step 9 — Find Email Addresses

```python
import re

text = '''
Contact us at support@test.com
or admin@company.com.au
'''

emails = re.findall(r"\S+@\S+", text)

print(emails)
```

## Step 10 — Find Australian Phone Numbers

```python
import re

text = "Call 0412 555 111 or 0399998888"

phones = re.findall(r"\d{4}\s\d{3}\s\d{3}", text)

print(phones)
```

## Step 11 — Extract Parts of Text

```python
import re

text = "Name: Chris"

result = re.search(r"Name:\s(\w+)", text)

print(result.group(1))
```

## Step 12 — Replace Words

```python
import re

text = "I love cats"

result = re.sub("cats", "dogs", text)

print(result)
```

## Step 13 — Split Using Regex

```python
import re

text = "apple,banana;orange grape"

result = re.split(r"[,; ]", text)

print(result)
```

## Step 14 — Why We Use Raw Strings

```python
pattern = r"\d+"

print(pattern)
```

## Step 15 — Extract All Years

```python
import re

text = '''
1999 was important.
In 2025 AI became mainstream.
2030 may look very different.
'''

years = re.findall(r"\d{4}", text)

print(years)
```

## Step 16 — Validate a Simple Email

```python
import re

email = "test@example.com"

pattern = r"\S+@\S+\.\S+"

if re.fullmatch(pattern, email):
    print("Valid")
else:
    print("Invalid")
```

## Step 17 — Find Words Starting with A

```python
import re

text = "Apple Apricot Banana Avocado"

words = re.findall(r"\bA\w+", text)

print(words)
```

## Step 18 — Regex is Pattern Matching

```python
import re

text = "The years are 1999 and 2025"

matches = re.findall(r"\d{4}", text)

print(matches)
```

## Step 19 — Greedy Matching

```python
import re

text = "<h1>Title</h1><p>Hello</p>"

result = re.findall(r"<.*>", text)

print("Greedy:", result)

result = re.findall(r"<.*?>", text)

print("Non-greedy:", result)
```

## Step 20 — Summary of Common Patterns

```python
summary = {
    r"\d+": "One or more digits",
    r"\w+": "One or more word characters",
    r"\s": "Whitespace",
    ".": "Any character",
    ".*": "Anything",
    "^": "Start",
    "$": "End",
    "()": "Capture group",
    "[]": "Character list",
    "{}": "Exact count"
}

for k, v in summary.items():
    print(f"{k} -> {v}")
```
